```python
# Partition Function for Quicksort
def partition(TheArray, low, high):

  # Element to be placed at right position
  pivot = TheArray[high]

  # Index of smaller element
  i = (low-1) # -1 to start

  for j in range(low, high):
    # if current element is smaller than the pivot
    if (TheArray[j] <= pivot):

      i+=1
      
      TheArray[i], TheArray[j] = TheArray[j], TheArray[i]

  TheArray[i+1], TheArray[high] = TheArray[high], TheArray[i+1]

  #print(TheArray)

  return i+1

# Quicksort Function
def quicksort(TheArray, low, high):

  if (low < high):

    #print('Quicksort Layer')
    partitionIndex = partition(TheArray, low, high)
    #print('partitionIndex: ' + str(partitionIndex))

    # Recursively call quicksort on the low-middle section
    quicksort(TheArray, low, partitionIndex-1)

    # Recursively call quicksort on the middle-high section
    quicksort(TheArray, partitionIndex+1, high)

  return TheArray
```

```python
import random


def generateRandomList(numberOfElements, low, high):

  randomList = []

  for i in range(0, numberOfElements):

   entry = random.randint(low, high)
   randomList.append(entry)

  return randomList

  
```

```python
import math

def printFunction(SortedArray):

  x = math.ceil(len(SortedArray)/25)
  startingIndex = 0
  endingIndex = 25

  for i in range(0, x):
 
    print(str(SortedArray[startingIndex : endingIndex]))
    startingIndex += 25
    endingIndex += 25
```

```python
import time

#####  Part a  ######
arrayA = [10,80,3,19,14,7,5,12]
print('Unsorted Array Part a: ')
printFunction(arrayA)

# The starting index
low = 0
# The last index
high = (len(arrayA) - 1)

starta = time.time()
sortedArray = quicksort(arrayA, low, high)
enda = time.time()
runtimea = enda - starta


#print('Sorted Array Part a: ' + str(sortedArray))
print('Sorted Array Part a: ')
printFunction(sortedArray)
print('Sorting Runtime: ' + str(runtimea*1000) + ' milliseconds' + '\n')



#####  Part b  ######
array100 = generateRandomList(100, 1, 1000)
print('Unsorted Array Part b: ')
printFunction(array100)

# The starting index
low = 0
# The last index
high = (len(array100) - 1)

startb = time.time()
sortedArrayb = quicksort(array100, low, high)
endb = time.time()
runtimeb = endb - startb
print('Sorted Array Part b: ')
printFunction(sortedArrayb)
print('Sorting Runtime: ' + str(runtimeb*1000) + ' milliseconds' + '\n')



#####  Part c  ######
array1000 = generateRandomList(1000, 1, 1000)
print('Unsorted Array Part c: ')
printFunction(array1000)

# The starting index
low = 0
# The last index
high = (len(array1000) - 1)

startc = time.time()
sortedArrayc = quicksort(array1000, low, high)
endc = time.time()
runtimec = endc - startc
print('Sorted Array Part c: ')
printFunction(sortedArrayc)
print('Sorting Runtime: ' + str(runtimec*1000) + ' milliseconds' + '\n')
```
